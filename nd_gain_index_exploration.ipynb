###**About the dataset**

The dataset chosen for this exploration is the Notre Dame Global Adaptation Index (ND-GAIN). It is a program of the University of Notre Dame’s Environmental Change Initiative. The ND-GAIN ranks the climate adaptation performance for 177 countries over the last 17 years. One of ND-GAIN’s goals is to assist decision-makers in the public and private sectors to gain a better understanding of the climate adaptation. This can also help decision makers to identify and prioritize adaptation measures to allocate investment most effectively and build resilience to climate change.


## **1.2 ND-GAIN Indicators**

The following indicators of the ND-GAIN are to be used in this exploration: 

### **Vulnerability**
* Health - measured from the number of medical staffs, change of deaths from climate change induced diseases, slum population, dependency on external resource for health services, access to improved sanitation facilities
* Infrastructure - dependency on imported energy, electricity access, disaster preparedness, projection of sea level rise impacts, projection change of hydropower generation capacity, population living under 5m above sea level
* Human Habitat - quality of trade and transport-related infrasturucture, urban concentration, projected warm periord, age dependency ratio, paved roads, projected change of flood hazard
* Ecosystem Services - protected biomes, dependency on natural capital, projected change of biome distribution, ecological footprint, engagement in international environment conventions, projected change of marine biodiversity
* Food - agriculture capacity, food import dependency, projected population change, rural population, child malnutrition
* Water - fresh water withdrawal rate, access to reliable drinkin water, dam capacity, water ratio dependency, projected change of groundwater recharge

### **Readiness**
* Governance Readiness - the stability of society and institutional arrangements that contribute to investment risks, political stability and non-violence, control of corruption, regulatory quality, rule of law
* Social Readiness - social conditions that help society to make efficient and equitable use of investment and yield more benefit from the investment, social inequality, ICT infrastructure, education, innovation
* Economic Readiness - measured through overall investments of firms in the country to support green initiatives

We want to explore the indicators through a comparison of ideal and actual prioritization over the years. To assume an ideal rank, we want to look into the scores for 1995, which is the first year of which the data was collected. 

In this section, we present the initial scores from the vulnerability indicators in 1995 for three countries: Philippines, Indonesia, and Thailand. Initially, we assume that these are the vulnerability indicators that are expected to be improved by the countries for the next years. Assumming that these countries want to improve this over the years, here is the ranking of the vulnerability indicators that need improvement for all the countries:



```python
# Creating DataFrame of Ideal Rank per Country with reference to year 1995
v_ph_1995 = []
v_th_1995 = []
v_ind_1995 = []
r_ph_1995 = []
r_th_1995 = []
r_ind_1995 = []
vulnerability_indicators = [df_infra, df_health, df_eco, df_habitat, df_food, df_water]
readiness_indicators = [df_econ, df_social, df_gov]
# getting data from year 1995 per country per indicator
for category in vulnerability_indicators:
  v_ph_1995.append(category['1995']['Philippines'])
  v_th_1995.append(category['1995']['Thailand'])
  v_ind_1995.append(category['1995']['Indonesia'])
for category in readiness_indicators:
  r_ph_1995.append(category['1995']['Philippines'])
  r_th_1995.append(category['1995']['Thailand'])
  r_ind_1995.append(category['1995']['Indonesia'])

# creating a dictionary showing the score to its indicator
v_indicators = ['Infrastructure', 'Health', 'Ecosystem', 'Habitat', 'Food', 'Water']
r_indicators = ['Economic Readiness', 'Social Readiness', 'Governance Readiness']
v_ph_ideal = {'Category': v_indicators, 'Score': v_ph_1995}
v_th_ideal = {'Category': v_indicators, 'Score': v_th_1995}
v_ind_ideal = {'Category': v_indicators, 'Score': v_ind_1995}
r_ph_ideal = {'Category': r_indicators, 'Score': r_ph_1995}
r_th_ideal = {'Category': r_indicators, 'Score': r_th_1995}
r_ind_ideal = {'Category': r_indicators, 'Score': r_ind_1995}

# converting dictionary to dataframe 
df_ph_v_ideal = pd.DataFrame(v_ph_ideal)
df_th_v_ideal = pd.DataFrame(v_th_ideal)
df_ind_v_ideal = pd.DataFrame(v_ind_ideal)
df_ph_r_ideal = pd.DataFrame(r_ph_ideal)
df_th_r_ideal = pd.DataFrame(r_th_ideal)
df_ind_r_ideal = pd.DataFrame(r_ind_ideal)
```

```python
df_ph_v_ideal # highest score is ranked first (most in need for improvement)

#to visualize the ideal vulnerability indicators that needs improvement
ax = sns.barplot(x="Category", y="Score", data=df_ph_v_ideal)
ax.set_title('Initial Vulnerability Indicator Ranking of the Philippines') #add title
```

As shown above in the case of the Philippines, notice that **Habitat** needs the most improvement from 1995, followed by Water, Food, Health, Infrastrucutre, and Ecosystem respecitvely.

Then we look into the actual progress throughout the years in which the mean rate of change per country per indicator was calculated. The ranking for vulnerability meant that the more negative the number was, the better its improvement has been throughout the years.

```python
# Creating DataFrame of Actual Rank per Country using mean rate of change
v_ph_mean_changes = []
v_th_mean_changes = []
v_ind_mean_changes = []
r_ph_mean_changes = []
r_th_mean_changes = []
r_ind_mean_changes = []

# Computing Mean Rate of Change
# For Vulnerability Indicators
vulnerability_indicators = [df_infra, df_health, df_eco, df_habitat, df_food, df_water]
for category in vulnerability_indicators:
  category = category.T
  category["Rate of Change (PH)"] = category["Philippines"].shift(-1) - category["Philippines"]
  v_ph_mean_changes.append(category['Rate of Change (PH)'].mean())
  category["Rate of Change (TH)"] = category["Thailand"].shift(-1) - category["Thailand"]
  v_th_mean_changes.append(category['Rate of Change (TH)'].mean())
  category["Rate of Change (IND)"] = category["Indonesia"].shift(-1) - category["Indonesia"]
  v_ind_mean_changes.append(category['Rate of Change (IND)'].mean())
# For Readiness Indicators
readiness_indicators = [df_econ, df_social, df_gov]
for category in readiness_indicators:
  category = category.T
  category["Rate of Change (PH)"] = category["Philippines"].shift(-1) - category["Philippines"]
  r_ph_mean_changes.append(category['Rate of Change (PH)'].mean())
  category["Rate of Change (TH)"] = category["Thailand"].shift(-1) - category["Thailand"]
  r_th_mean_changes.append(category['Rate of Change (TH)'].mean())
  category["Rate of Change (IND)"] = category["Indonesia"].shift(-1) - category["Indonesia"]
  r_ind_mean_changes.append(category['Rate of Change (IND)'].mean())

# Creating dictionaries with indicators as keys and mean changes as values
v_indicators = ['Infrastructure', 'Health', 'Ecosystem', 'Habitat', 'Food', 'Water']
r_indicators = ['Economic Readiness', 'Social Readiness', 'Governance Readiness']
v_ph_changes = {'Category': v_indicators, 'Mean rate of change': v_ph_mean_changes}
v_th_changes = {'Category': v_indicators, 'Mean rate of change': v_th_mean_changes}
v_ind_changes = {'Category': v_indicators, 'Mean rate of change': v_ind_mean_changes}
r_ph_changes = {'Category': r_indicators, 'Mean rate of change': r_ph_mean_changes}
r_th_changes = {'Category': r_indicators, 'Mean rate of change': r_th_mean_changes}
r_ind_changes = {'Category': r_indicators, 'Mean rate of change': r_ind_mean_changes}

# dictionaries to dataframes
df_ph_v_actual = pd.DataFrame(v_ph_changes)
df_th_v_actual  = pd.DataFrame(v_th_changes)
df_ind_v_actual = pd.DataFrame(v_ind_changes)
df_ph_r_actual = pd.DataFrame(r_ph_changes)
df_th_r_actual = pd.DataFrame(r_th_changes)
df_ind_r_actual= pd.DataFrame(r_ind_changes)
```

Now that we know the initial ranking of the indicators. Let's compare this to what indicators did the Philippines have **actually improved over the years** (starting from 1995 to 2018):

```python
df_ph_v_actual # most negative is ranked first (most improved or prioritized over the years)

ax = sns.barplot(x="Category", y="Mean rate of change", data=df_ph_v_actual.sort_values('Mean rate of change'))
ax.set_title('Vulnerability Indicators Prioritized by the Philippines Over the Years') #add title
```

This is a bit concerning that indicators were not improved as suggested from the initial ranking that were released in 1995. Which begs the question: **do countries care about the ND-Gain Scores?**

After creating the bar charts, it came to mind that the purpose of the graphs could be visualized in another way. Creating proportional area charts that line up vertically may be easier to show ranking. Also, having an ideal vs actual visualization may allow comparisons to be done easier.

```python
# proportion size variable (will change)
v_median = 0.35
```

```python
# preparing dataframes for plotting
ideal_group = [df_ph_v_ideal, df_th_v_ideal, df_ind_v_ideal]
for df in ideal_group:
  df.sort_values(by='Score', inplace=True) # to organize for ranking
  df['Rank(I)'] = range(1,7) # display ranking
  df['ideal'] = 4 # for placement in x-axis
  df['bubble_size'] = abs(df['Score']-v_median)/(df['Score']-v_median).max()*10000 # proportional size of circle

actual_group = [df_ph_v_actual, df_th_v_actual, df_ind_v_actual]
for df in actual_group:
  df.sort_values(by='Mean rate of change', ascending=False, inplace=True)
  df['Rank(A)'] = range(1,7)
  df['actual'] = 8
  df['bubble_size'] = abs(df['Mean rate of change'])/abs(df['Mean rate of change'].min())*10000
```

```python
# sample of dataframe
# rank is only for plotting purposes but actually means the opposite
df_ph_v_ideal
```

```python
# lists for iterating through loops for plotting
titles = ["(PH)", "(TH)", "(IND)"]
filenames = ['ph_v.png', 'th_v.png', 'ind_v.png']

# color distinction for plotting
color_dict = {'Habitat': 'purple', 'Ecosystem': 'green', 'Infrastructure': 'grey', 'Health': 'red', 'Food': 'orange', 'Water':'blue'}
```

```python
# plotting vulnerability scores and rates
for i, a, title, filename in zip(ideal_group, actual_group, titles, filenames):
  fig, ax = plt.subplots(1,1, figsize=(10, 15), dpi=80)
  ax.scatter(x=i['ideal'],
            y=i['Rank(I)'],
            c=i['Category'].map(color_dict),
            s=i['bubble_size'])
  ax.scatter(x=a['actual'],
            y=a['Rank(A)'],
            c=a['Category'].map(color_dict),
            s=a['bubble_size'])

  # add a condition to recognize that a score is above the median of vulnerability scores

  # include labels per point
  for p1, c, score in zip(i['Rank(I)'], i['Category'], i['Score']):
    ax.text(0, p1, str(round(score, 4)) + ', \n' + c , horizontalalignment='left', verticalalignment='center', fontdict={'size':18, 'weight': 600})

  for p2, c, rate in zip(a['Rank(A)'], a['Category'], a['Mean rate of change']):
    ax.text(9.5, p2, "{:.4%}".format(rate) + ', \n' + c, horizontalalignment='left', verticalalignment='center', fontdict={'size':18, 'weight': 600})

  # Ideal vs Actual titles
  ax.text(4, 6.6, 'IDEAL', horizontalalignment='center', verticalalignment='center', fontdict={'size':18, 'weight':700})
  ax.text(6, 6.6, 'vs', horizontalalignment='center', verticalalignment='center', fontdict={'size':18, 'weight':700})
  ax.text(8, 6.6, 'ACTUAL', horizontalalignment='center', verticalalignment='center', fontdict={'size':18, 'weight':700})
  # hide ranking labels
  plt.yticks([])
  plt.xticks([])
  # Layout
  ax.set_title("VULNERABLE SECTORS " + title + " VS AVE RATE OF IMPROVEMENT \n IN 23 YEARS", fontdict={'size':22})

  ax.set_xlim([0,12])
  ax.set_ylim([0, 7])

  # remove border lines
  ax.spines["top"].set_alpha(.0)
  ax.spines["bottom"].set_alpha(.0)
  ax.spines["right"].set_alpha(.0)
  ax.spines["left"].set_alpha(.0)
  # save figure
  plt.savefig(filename)
```

##4.1 Summary of Analysis
In terms of vulnerability scores, it was found through data analysis that habitat was the most vulnerable sector for the Philippines while food was the most vulnerable sector for both Thailand and Indonesia based on the data from 1995. However, through analyzing the average rate of change for the scores of each sector from 1995 to 2018, it was observed that all three countries focused on the infrastructure sector as this was seen with the greatest decrease in terms of rate of change. 

In terms of readiness scores, social readiness was found to have the lowest readiness score among all sectors for the three countries based on the data from 1995. Through analyzing the average rate of change for the scores of each sector from 1995 to 2018, it was observed that both Thailand and the Philippines gave priority to social readiness which was found to be the least ready sector in 1995. However, Indonesia gave priority to economic readiness over social readiness given that it was observed to have a more positive average rate of change.

##4.2 Recommendations

####Use of Rate of Change

For the analysis done in this project, the average rate of change was used as a metric to analyze the progress of the country in terms of their vulnerability and readiness scores from 1995 to 2018. The average rate of change is a metric not often used in these types of analyses; however, it can be very useful in giving a good visualization of the progress each country has made with respect to the individual indicators in the 23-year span. Therefore, one of the recommendations for this project is for future analyses to make use of the average rate of change as a metric in analyzing similar datasets such as comparing progress.


####Policy Recommendations
Another recommendation from this project is to make use of the ideal ranking in comparison with the actual scores (rate of change) in order to give priority to the sectors that are most vulnerable. The same recommendation can be made in terms of the sectors that fall under readiness wherein the sectors that are least ready would be given priority. Priority in terms of policy implementation and investment should be given to the most vulnerable and least ready sectors.
